```python
def resolucao():
    n,i,B,l,e,V = symbols('n,i,B,l,e,V')
    print("Partindo da equação de tensao em uma espira temos")
    display(Eq(n,(B*i)/(V*l*e)))
    print("Isolando a diferença de potencial V temos:")
    display(Eq(V,(B*i)/(n*l*e)))
    n1,i1,B1,l1,e1,V1 = symbols('8.47*10__28,23,0.65,150*10__-6,1.6*10__-19,V')
    display(Eq(V,(B1*i1)/(n1*e1*l1)))
    res = symbols('7.4*10__-6')
    display(Eq(V,res*V))
    
```

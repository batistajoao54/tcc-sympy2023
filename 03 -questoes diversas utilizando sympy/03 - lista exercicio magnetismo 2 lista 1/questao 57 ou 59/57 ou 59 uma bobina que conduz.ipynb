```python
from sympy import *
```

```python
def resultado():
    µ,i,A = symbols('µ,i,A')
    print("Momento dipolar magnetico de uma bobina é dado por")
    display(Eq(µ,i*A))
    print("Onde A é a area, logo (0.3 x 0.4)/2 em metros")
    display(Eq(A,0.06))
    display(Eq(i,5))
    print('O item "a" tem como resposta')
    r = symbols('0.06*5')
    display(Eq(µ,r))
    res = symbols('0.30Am__2')
    display(Eq(µ,res))
    print()
    print("O item 'b' tem como torque a seguinte equação")
    τ,θ,B = symbols('τ,θ,B')
    display(Eq(τ,µ*B*sin(θ)))
    print("sin(90°) = 1, logo temos")
    display(Eq(τ,µ*B))
    res1,res2 = symbols('80*10__-3,T')
    display(Eq(τ,res1*res2*res))
    fim,o = symbols('2.4*10__-2,Nm')
    display(Eq(τ,fim*o))
```

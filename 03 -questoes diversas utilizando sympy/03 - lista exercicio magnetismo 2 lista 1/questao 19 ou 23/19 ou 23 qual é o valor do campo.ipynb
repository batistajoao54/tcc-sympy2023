```python
from sympy import *
```

```python
def resultado():
    B,me,v,e,r = symbols('B,m_e,v,e,r')
    print("Partindo da equação da trajetoória circula")
    display(Eq(r,me*v/(e*B)))
    print("Isolando o campo magnetico B temos")
    display(Eq(B,me*v/(r*e)))
    print("Dados")
    massa,kg = symbols('9.11*10__-31,kg')
    eletron,C = symbols('1.60*10__-19,C')
    velocidade,ms = symbols('1.30*10__6,m/s')
    raio,m = symbols('0.350,m')
    display(Eq(v,velocidade*ms))
    display(Eq(r,raio*m))
    display(Eq(me,massa*kg))
    display(Eq(e,eletron*C))
    print("Substituindo temos a seguinte resposta para B")
    display(Eq(B,massa*velocidade/(raio*eletron)))
    res,res1 = symbols('2.11*10__-5,T')
    display(Eq(B,res*res1))
```

# questao

```python
resultado()
```

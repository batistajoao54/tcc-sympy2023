```python
from sympy import*
```

```python
def resultado():
    Fb,Fp,i,L,m,g,mg,B = symbols('F_B,F_P,i,L,m,g,mg,B ')
    print("a magnitude da força magnética é dada por:")
    display(Eq(Fb,i*L*B))
    display(Eq(Fp,mg))
    print("Igualando as duas forças temos")
    display(Eq(Fb,Fp))
    display(Eq(B*L*i,mg))
    print("Isolando a corrente 'i'")
    display(Eq(i,mg/(B*L)))
    print("Dados")
    display(Eq(m,0.013))
    display(Eq(L,0.62))
    display(Eq(B,0.44))
    display(Eq(g,9.8))
    num,deno = symbols('0.013*9.8,0.44*0.62')
    print("Substituindo na equação")
    display(Eq(i,num/deno))
    res = symbols('0.467A')
    display(Eq(i,res))
```

# Questao

```python
resultado()
```

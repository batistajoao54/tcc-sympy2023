```python
from sympy import *
```

```python
def resposta():
    E,v,B,m,e,V,Fm,Fe,q,Fr = symbols('E,v,B,m,e,V,F_B,F_E,q,F_r')
    print("Força Magnetica")
    display(Eq(Fm,q*v*B))
    print('Força Eletrica')
    display(Eq(Fe,E*q))
    print("Força resultante é")
    display(Eq(Fm+Fe,(q*v*B)+(E*q)))
    display(Eq(Fr,q*(E+v*B)))
    display(Eq(Fr,e*(E+v*B)))
    print("Força resultando é igual a ZERO, então temos")
    display(Eq(E,B*v))
    print("Agora precisamos desenvolver o v, que é dado por")
    display(Eq(e*V,(m*v**2)/2))
    display(Eq(v**2,(2*e*V)/m))
    r,T,C = symbols('2Ve,T,C')
    display(Eq(v,sqrt(r/m)))
    print("Pronto agora so precisamos substituir em E = Bv")
    display(Eq(E,B*sqrt(r/m)))
    print("Dados")
    display(Eq(B,1.2*T))
    display(Eq(V,10000*V))
    s,kg,es = symbols('9.99x10__-27,kg,1.60x10__-19')
    display(Eq(m,s*kg))
    display(Eq(e,es*C))
    print('So precisamos substituir')
    ss,rr = symbols('9.99x10__-27,10000*2')
    display(Eq(E,1.2*sqrt((es*rr)/ss)))
    res,sei = symbols('6.8*10__5,V/m')
    display(Eq(E,res*sei))
```

# Questão

```python
resposta()
```

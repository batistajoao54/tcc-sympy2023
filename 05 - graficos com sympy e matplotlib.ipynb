# Usaremos a equação dos gases ideais

```python
#criando simbolos e a equação
P, V, T, R, n = sympy.symbols('P V T R n', posivite=True)

ideal_gas_law = sympy.Eq(P*V, n*R*T)
display(ideal_gas_law)
```

## Deixando claro o papel de cada símbolo:

##### P: pressão em atmosferas (atm)
##### V: volume em litros
##### n: quantidade de matéria em mol
##### R: constante dos gases em atm.l/(mol.K)
##### T: temperatura em Kelvin
##### Vamos resolver tal equação para o volume V:

```python
#resolvendo exolando V
display(sympy.solve(ideal_gas_law, V)) #retorma uma lista com as respostas

display(sympy.solve(ideal_gas_law, V)[0])

#podemos melhorar essa resolução

print('Melhorando a equação:')

display(sympy.Eq(V,sympy.solve(ideal_gas_law, V)[0]))
#como ao usar a funçao solver devolve uma lista assimpodemos usar o fatiamento para tirar apenas um valor de dentro 
#dessa lista
```

```python
#cologando essa equação para uma variavel
volume = sympy.Eq(V,sympy.solve(ideal_gas_law, V)[0])
display(volume)
```

### 'R' e 'T' e 'n' sendo constante teremos


```python
#atribuindo valores as respectivas letras teremos
gas_constant = 0.082057366080960
volume_molar_25_celsius = volume.subs({R: gas_constant, n: 1, T: 298.15})
display(volume_molar_25_celsius)

#para constuir o grafico nao podemos igualar a equação que queremos
#entao irei modificar outra variavel porem nao vai ter a igualdade do "V"
volume2 = sympy.solve(ideal_gas_law, V)[0]
display('TESTE', volume2)
volume_molar = volume2.subs({R: gas_constant, n: 1, T: 298.15})
display("uaremos esse no grafico", volume_molar)
```

## Agora podemos construir um grafico com variação em 'V' e em 'P'

```python
#Gráfico com sympy
plot(volume_molar, P)
```

```python
#melhorando o dominio do grafico
plot(volume_molar, (P, 0.1, 4))
```

```python
#podemos alterar os limites do eixo y tbem usando a função ylim = (inicio, fim)
plot(volume_molar, (P, 0.1, 4), ylim=(0, 80))
```

E se quisermos outra curva no gráfico? Por exemplo, para 100 °C (373,15 K)? Simples, basta criar a expressão e passá-la para o método plot:

```python
volume_molar_100_celsius = volume2.subs({R: gas_constant, n: 1, T: 373.15})
display(volume_molar_100_celsius)
```

```python
plot(volume_molar, volume_molar_100_celsius, (P, 0.1, 4), ylim=(0, 80),)
```

```python
temperatures_celsius = (-100, 0, 25, 100, 500, 1000)
temperatures_kelvin = tuple(t + 273.15 for t in temperatures_celsius)
```

```python
temperatures_kelvin
```

### Agora, pense. Queremos seis gráficos, precisamos dos dados de cada um. Vamos criar uma lista vazia para armazenas esses dados. E esses dados precisam ser extraídos de expressões do SymPy com o get_points. Podemos criar cada expressão e extrair, mas não precisamos efetivamente criar uma expressão para cada temperatura, podemos reaproveitar a mesma variável, não há interesse em objetos para cada expressão. Basicamente, descrevemos um loop for:

```python
plots = []

for temperature in temperatures_kelvin:
    expr = volume2.subs({R: gas_constant, n: 1, T: temperature})
    p = plot(expr, (P, 0.1, 4), show=False)
    data = p[0].get_points()
    plots.append(data)
```

```python
fig, ax = plt.subplots(figsize=(12, 8), facecolor=(1, 1, 1))

for i, p in enumerate(plots):
    ax.plot(*p, label=f'T = {temperatures_celsius[i]} °C', linewidth=3)
    ax.set_ylim(0, 80)
    ax.legend(fontsize=14)
    ax.set_ylabel('Volume / liter', fontsize=16)
    ax.set_xlabel('Pressure / atm', fontsize=16)

ax.set_title("Molar volume - ideal gas", fontsize=18)

plt.show()
```
